```python
from sklearn.datasets import load_iris
import pandas as pd

#Cargamos el dataset IRIS
iris4= load_iris()

#Lo convertimos a una Dataframe de pandas
iris4_df = pd.DataFrame(data=iris4.data, columns=iris4.feature_names)
#Agragemos la columna de especie
iris4_df['species']= iris4.target
#Mostramos las primeras filas del Dataframe
iris4_df.head()
```

## Preprocesamiento

```python
X= iris4_df.drop('species', axis=1) #quitamos la columna de species
y= iris4_df['species']
```

```python
import numpy as np
X_media= X - np.mean(X, axis=0) #np.mean es la media de la base de datos
X_std= X_media/np.std(X, axis=0) #np.std es la desviaci√≥n estandar de la base de datos
cov_matrix= np.cov(X_std.T) #np.cov es la matriz de covarianza de la base de datos
cov_matrix
```

```python
eigen_values, eigen_vectors= np.linalg.eig(cov_matrix) #np.linalg.eig es la funci√≥n que calcula los valores y vectores propios de una matriz
eigen_values_index= np.argsort(eigen_values)[::-1] #np.argsort es la funci√≥n que ordena los valores propios de menor a mayor
eigen_values_chamber= eigen_values[eigen_values_index]

eigen_vectors_chamber= eigen_vectors[:, eigen_values_index]

eigen_values_chamber
```

```python
eigen_vectors_chamber
```

```python
eigen_subset= eigen_vectors_chamber[:, 0:2] #seleccionamos los dos primeros vectores propios
eigen_subset
```

```python
X_red= np.dot(eigen_subset.transpose(), X_std.transpose()).transpose() #np.dot es la funci√≥n que multiplica dos matrices
X_red
```

```python
principal_df= pd.DataFrame(X_red, columns=['PC1', 'PC2']) #creamos un dataframe con los dos vectores propios
principal_df
```

```python
principal_df= pd.concat([principal_df, iris4_df[['species']]], axis=1) #concatenamos el dataframe con la columna de species
principal_df
```

```python
import plotly.express as px

fig= px.scatter(principal_df, x='PC1', y='PC2', color='species') #creamos un gr√°fico de dispersi√≥n con los dos vectores propios
fig.show()
```
